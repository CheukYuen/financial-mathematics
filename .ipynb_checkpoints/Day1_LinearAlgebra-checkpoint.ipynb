### ‰ªäÂ§©ÁöÑÂ≠¶‰π†ÈáåÁ®ãÁ¢ë

‚úÖ ÁêÜËß£Êî∂ÁõäÂêëÈáè Œº ‰∏éÂçèÊñπÂ∑ÆÁü©Èòµ Œ£ ÁöÑÈáëËûçÂê´‰πâ

‚úÖ ÈÄöËøáÊ®°ÊãüÊï∞ÊçÆÂÆåÊàêÔºöÁõ∏ÂÖ≥ÊÄßÁªìÊûÑ ‚Üí Œº/Œ£/œÅ ËÆ°ÁÆó ‚Üí Âá†‰ΩïÂèØËßÜÂåñ  
‚úÖ ÁêÜËß£ÁâπÂæÅÂàÜËß£‰∏é‰∏ªÈ£éÈô©ÊñπÂêëÁöÑÂÖ≥Á≥ª  
‚úÖ ËÆ°ÁÆóÊäïËµÑÁªÑÂêàÈ£éÈô©Ê¢ØÂ∫¶Âπ∂ËøõË°åÈ£éÈô©‰ºòÂåñÂÆûÈ™å  
‚úÖ ÂÆåÊàê5‰∏™Âä®ÊâãÁªÉ‰π†ÔºåÂ∑©Âõ∫ÁêÜËÆ∫Áü•ËØÜ


````python
# ‰æùËµñÊ£ÄÊü•‰∏éÂÆâË£Ö
import sys
import subprocess
import importlib

def check_and_install_package(package_name, import_name=None):
    """Ê£ÄÊü•Âπ∂ÂÆâË£ÖÁº∫Â§±ÁöÑÂåÖ"""
    if import_name is None:
        import_name = package_name
    
    try:
        importlib.import_module(import_name)
        print(f"‚úÖ {package_name} Â∑≤ÂÆâË£Ö")
        return True
    except ImportError:
        print(f"‚ö†Ô∏è  {package_name} Êú™ÂÆâË£ÖÔºåÊ≠£Âú®Â∞ùËØïÂÆâË£Ö...")
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", package_name])
            print(f"‚úÖ {package_name} ÂÆâË£ÖÊàêÂäü")
            return True
        except subprocess.CalledProcessError:
            print(f"‚ùå {package_name} ÂÆâË£ÖÂ§±Ë¥•ÔºåÂ∞Ü‰ΩøÁî®Êõø‰ª£ÊñπÊ°à")
            return False

# Ê£ÄÊü•Ê†∏ÂøÉ‰æùËµñ
required_packages = [
    ("numpy", "numpy"),
    ("pandas", "pandas"), 
    ("matplotlib", "matplotlib"),
    ("scipy", "scipy"),
    ("seaborn", "seaborn")  # ÂèØÈÄâÔºåÁî®‰∫éÊõ¥Â•ΩÁöÑÁÉ≠ÂäõÂõæ
]

seaborn_available = True
for package, import_name in required_packages:
    if package == "seaborn":
        seaborn_available = check_and_install_package(package, import_name)
    else:
        check_and_install_package(package, import_name)

if not seaborn_available:
    print("üìä Â∞Ü‰ΩøÁî® matplotlib Êõø‰ª£ seaborn ËøõË°åÁÉ≠ÂäõÂõæÁªòÂà∂")
````

## SetupÔΩúÁéØÂ¢É‰∏éÂ∑•ÂÖ∑


````python
# ÂØºÂÖ•Ê†∏ÂøÉÂ∫ì
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import linalg
import os
from pathlib import Path

# ËÆæÁΩÆÈöèÊú∫ÁßçÂ≠êÁ°Æ‰øùÁªìÊûúÂèØÂ§çÁé∞
rng = np.random.default_rng(42)

# ÈÖçÁΩÆÁªòÂõæÈ£éÊ†º
if seaborn_available:
    import seaborn as sns
    sns.set_style("whitegrid")
    sns.set_palette("husl")
    print("üìä ‰ΩøÁî® seaborn ‰∏ì‰∏öÁªòÂõæÈ£éÊ†º")
else:
    plt.style.use('default')
    print("üìä ‰ΩøÁî® matplotlib ÈªòËÆ§ÁªòÂõæÈ£éÊ†º")

# ËÆæÁΩÆ‰∏≠ÊñáÂ≠ó‰ΩìÊîØÊåÅ
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# ÂÆö‰πâÂ∏∏Èáè
ASSETS = ["CashProxy", "Bond", "Equity", "Gold"]
n_assets = len(ASSETS)

# Âπ¥ÂåñÂèÇÊï∞
mu_annual = np.array([0.02, 0.045, 0.08, 0.03])        # Âπ¥ÂåñÊúüÊúõÊî∂Áõä
sigma_annual = np.array([0.005, 0.06, 0.18, 0.12])     # Âπ¥ÂåñÊ≥¢Âä®

# Áõ∏ÂÖ≥Á≥ªÊï∞Áü©ÈòµÔºàÈáëËûçÁõ¥ËßâÔºöÂÄ∫Âà∏‰∏éËÇ°Á•®Ë¥üÁõ∏ÂÖ≥ÔºåÈªÑÈáë‰∏éËÇ°Á•®‰ΩéÁõ∏ÂÖ≥Ôºâ
rho = np.array([
    [1.00,  0.10, -0.05,  0.00],  # CashProxy
    [0.10,  1.00, -0.30,  0.10],  # Bond  
    [-0.05,-0.30,  1.00,  0.15],  # Equity
    [0.00,  0.10,  0.15,  1.00],  # Gold
])

# Êó∂Èó¥ÂèÇÊï∞
TRADING_DAYS = 252  # Âπ¥‰∫§ÊòìÊó•Êï∞
T = 750             # Ê†∑Êú¨ÈïøÂ∫¶ÔºàÁ∫¶3Âπ¥Ôºâ
OUTDIR = "./artifacts"

# ÂàõÂª∫ËæìÂá∫ÁõÆÂΩï
Path(OUTDIR).mkdir(exist_ok=True)
print(f"üìÅ ËæìÂá∫ÁõÆÂΩï: {OUTDIR}")

print("üîß ÁéØÂ¢ÉÈÖçÁΩÆÂÆåÊàêÔºÅ")
print(f"üìä ËµÑ‰∫ßÁ±ªÂà´: {ASSETS}")
print(f"üìà Ê†∑Êú¨ÈïøÂ∫¶: {T} ‰∏™‰∫§ÊòìÊó•")
print(f"üéØ Âπ¥ÂåñÊúüÊúõÊî∂Áõä: {mu_annual}")
print(f"üìâ Âπ¥ÂåñÊ≥¢Âä®Áéá: {sigma_annual}")
````

## Data SimulationÔΩúÊ®°ÊãüÂ§öËµÑ‰∫ßÊî∂Áõä

Áé∞Âú®Êàë‰ª¨Â∞Ü‰ΩøÁî®È¢ÑËÆæÁöÑÁõ∏ÂÖ≥ÁªìÊûÑÊù•ÊûÑÈÄ†ÂçèÊñπÂ∑ÆÁü©ÈòµÔºåÂπ∂ÁîüÊàêÂ§öËµÑ‰∫ßÁöÑÊ®°ÊãüÊî∂ÁõäÊï∞ÊçÆ„ÄÇ


````python
# Ê≠•È™§1: Â∞ÜÂπ¥ÂåñÂèÇÊï∞ËΩ¨Êç¢‰∏∫Êó•Â∫¶ÂèÇÊï∞
mu_daily = mu_annual / TRADING_DAYS
sigma_daily = sigma_annual / np.sqrt(TRADING_DAYS)

print("üìä Êó•Â∫¶ÂèÇÊï∞ËΩ¨Êç¢:")
print(f"   ÊúüÊúõÊî∂Áõä: {mu_daily}")
print(f"   Ê≥¢Âä®Áéá: {sigma_daily}")

# Ê≠•È™§2: ÊûÑÈÄ†ÂçèÊñπÂ∑ÆÁü©Èòµ Œ£ = D * œÅ * D
# ÂÖ∂‰∏≠ D = diag(œÉ) ÊòØÊ†áÂáÜÂ∑ÆÂØπËßíÁü©Èòµ
D = np.diag(sigma_daily)
Sigma = D @ rho @ D

print(f"\nüìà ÂçèÊñπÂ∑ÆÁü©Èòµ Œ£ (Êó•Â∫¶):")
print(Sigma)

# Ê≠•È™§3: Êï∞ÂÄºÁ®≥ÂÆöÊÄßÊ£ÄÊü•‰∏é‰øÆÊ≠£
def ensure_positive_definite(Sigma: np.ndarray, tol: float = 1e-8) -> np.ndarray:
    """
    Á°Æ‰øùÂçèÊñπÂ∑ÆÁü©ÈòµÊ≠£ÂÆö
    
    Args:
        Sigma: ÂçèÊñπÂ∑ÆÁü©Èòµ
        tol: ÂØπËßíÁ∫øÊâ∞Âä®ÂÆπÂøçÂ∫¶
        
    Returns:
        ‰øÆÊ≠£ÂêéÁöÑÊ≠£ÂÆöÂçèÊñπÂ∑ÆÁü©Èòµ
    """
    # ÂØπÁß∞Âåñ
    Sigma_sym = (Sigma + Sigma.T) / 2
    
    # Ê£ÄÊü•Ê≠£ÂÆöÊÄß
    try:
        np.linalg.cholesky(Sigma_sym)
        print("‚úÖ ÂçèÊñπÂ∑ÆÁü©ÈòµÂ∑≤Ê≠£ÂÆö")
        return Sigma_sym
    except np.linalg.LinAlgError:
        print("‚ö†Ô∏è  ÂçèÊñπÂ∑ÆÁü©ÈòµÈùûÊ≠£ÂÆöÔºåÊ∑ªÂä†ÂØπËßíÁ∫øÊâ∞Âä®...")
        # Ê∑ªÂä†ÂØπËßíÁ∫øÊâ∞Âä®Á°Æ‰øùÊ≠£ÂÆö
        eigenvals = np.linalg.eigvals(Sigma_sym)
        min_eigenval = np.min(eigenvals)
        if min_eigenval <= 0:
            Sigma_corrected = Sigma_sym + (tol - min_eigenval) * np.eye(Sigma_sym.shape[0])
            print(f"   Ê∑ªÂä†ÂØπËßíÁ∫øÊâ∞Âä®: {tol - min_eigenval:.2e}")
            return Sigma_corrected
        return Sigma_sym

Sigma = ensure_positive_definite(Sigma)

# Ê≠•È™§4: CholeskyÂàÜËß£ÁîüÊàêÊ®°ÊãüÊï∞ÊçÆ
try:
    L = np.linalg.cholesky(Sigma)
    print("‚úÖ CholeskyÂàÜËß£ÊàêÂäü")
    
    # ÁîüÊàêÊ†áÂáÜÊ≠£ÊÄÅÈöèÊú∫Êï∞
    Z = rng.standard_normal((T, n_assets))
    
    # Â∫îÁî®CholeskyÂèòÊç¢: R = Œº + L * Z
    returns = mu_daily + Z @ L.T
    
    print(f"üìä ÁîüÊàê {T} ‰∏™Êó•Â∫¶Êî∂ÁõäÊ†∑Êú¨")
    
except np.linalg.LinAlgError as e:
    print(f"‚ùå CholeskyÂàÜËß£Â§±Ë¥•: {e}")
    # ‰ΩøÁî®ÁâπÂæÅÂàÜËß£‰Ωú‰∏∫Â§áÈÄâÊñπÊ°à
    eigenvals, eigenvecs = np.linalg.eigh(Sigma)
    eigenvals = np.maximum(eigenvals, 1e-8)  # Á°Æ‰øùÈùûË¥ü
    L = eigenvecs @ np.diag(np.sqrt(eigenvals))
    Z = rng.standard_normal((T, n_assets))
    returns = mu_daily + Z @ L.T
    print("‚úÖ ‰ΩøÁî®ÁâπÂæÅÂàÜËß£Â§áÈÄâÊñπÊ°àÁîüÊàêÊï∞ÊçÆ")

# Ê≠•È™§5: ÂàõÂª∫DataFrameÂπ∂‰øùÂ≠ò
returns_df = pd.DataFrame(returns, columns=ASSETS)
returns_df.index.name = 'Date'

# ‰øùÂ≠ò‰∏∫parquetÊ†ºÂºè
returns_df.to_parquet(f"{OUTDIR}/day1_returns.parquet")
print(f"üíæ Êï∞ÊçÆÂ∑≤‰øùÂ≠òËá≥: {OUTDIR}/day1_returns.parquet")

# ÊòæÁ§∫Êï∞ÊçÆÊëòË¶Å
print(f"\nüìà Ê®°ÊãüÊï∞ÊçÆÊëòË¶Å:")
print(f"   ÂΩ¢Áä∂: {returns_df.shape}")
print(f"   ÂùáÂÄº: {returns_df.mean().values}")
print(f"   Ê†áÂáÜÂ∑Æ: {returns_df.std().values}")
print(f"\nÂâç5Ë°åÊï∞ÊçÆ:")
print(returns_df.head())
````

## Math Warm-upÔΩúÁ∫øÊÄß‰ª£Êï∞‰∏éÁªüËÆ°ÊúÄÂ∞èÂÖ®ÂÆ∂Ê°∂

ËÆ©Êàë‰ª¨ÂõûÈ°æÁ∫øÊÄß‰ª£Êï∞ÂíåÁªüËÆ°Â≠¶‰∏≠ÁöÑÊ†∏ÂøÉÊ¶ÇÂøµÔºåËøô‰∫õÊòØÁêÜËß£ËµÑ‰∫ßÈÖçÁΩÆÊï∞Â≠¶Âü∫Á°ÄÁöÑÂÖ≥ÈîÆ„ÄÇ


````python
# 4.1 ÂêëÈáè‰∏éÁü©ÈòµÂü∫Á°Ä
print("üî¢ ÂêëÈáè‰∏éÁü©ÈòµÂü∫Á°Ä")
print("=" * 50)

# ‰ªéÊï∞ÊçÆ‰∏≠ÊèêÂèñÂêëÈáèÁ§∫‰æã
sample_returns = returns_df.iloc[0].values  # Á¨¨‰∏ÄÂ§©ÁöÑÊî∂ÁõäÂêëÈáè
print(f"Êî∂ÁõäÂêëÈáèÁ§∫‰æã (Á¨¨‰∏ÄÂ§©): {sample_returns}")
print(f"ÂêëÈáèÂΩ¢Áä∂: {sample_returns.shape}")
print(f"ÂêëÈáèËΩ¨ÁΩÆ: {sample_returns.T}")
print(f"L2ËåÉÊï∞: {np.linalg.norm(sample_returns):.6f}")

# Áü©ÈòµÊìç‰Ωú
print(f"\nÁü©ÈòµÂΩ¢Áä∂: {returns_df.shape}")
print(f"Áü©ÈòµËΩ¨ÁΩÆÂΩ¢Áä∂: {returns_df.T.shape}")

# ÂÜÖÁßØ‰∏éËåÉÊï∞
def vector_operations_demo(x: np.ndarray, y: np.ndarray) -> None:
    """ÊºîÁ§∫ÂêëÈáèËøêÁÆó"""
    print(f"ÂêëÈáè x: {x}")
    print(f"ÂêëÈáè y: {y}")
    print(f"ÂÜÖÁßØ x¬∑y: {np.dot(x, y):.6f}")
    print(f"L2ËåÉÊï∞ ||x||: {np.linalg.norm(x):.6f}")
    print(f"Âçï‰ΩçÂêëÈáè x/||x||: {x / np.linalg.norm(x)}")

# ‰ΩøÁî®Ââç‰∏§Â§©ÁöÑÊï∞ÊçÆÊºîÁ§∫
x = returns_df.iloc[0].values
y = returns_df.iloc[1].values
vector_operations_demo(x, y)
````

````python
# 4.2 ÁªüËÆ°ÈáèËÆ°ÁÆó
print("\nüìä ÁªüËÆ°ÈáèËÆ°ÁÆó")
print("=" * 50)

# Ê†∑Êú¨ÂùáÂÄºÂêëÈáè: ŒºÃÇ = (1/T) Œ£ r_t
mu_hat = returns_df.mean().values
print(f"Ê†∑Êú¨ÂùáÂÄºÂêëÈáè ŒºÃÇ: {mu_hat}")

# Ê†∑Êú¨ÂçèÊñπÂ∑ÆÁü©Èòµ: Œ£ÃÇ = (1/(T-1)) (R - ŒºÃÇ)(R - ŒºÃÇ)^T
def compute_sample_covariance(returns: np.ndarray) -> np.ndarray:
    """
    ËÆ°ÁÆóÊ†∑Êú¨ÂçèÊñπÂ∑ÆÁü©Èòµ
    
    Args:
        returns: T√ón Êî∂ÁõäÁü©Èòµ
        
    Returns:
        n√ón ÂçèÊñπÂ∑ÆÁü©Èòµ
    """
    T, n = returns.shape
    mu = returns.mean(axis=0)
    centered = returns - mu  # ÂéªÂùáÂÄº
    return (centered.T @ centered) / (T - 1)

Sigma_hat = compute_sample_covariance(returns_df.values)
print(f"\nÊ†∑Êú¨ÂçèÊñπÂ∑ÆÁü©Èòµ Œ£ÃÇ:")
print(Sigma_hat)

# Áõ∏ÂÖ≥Áü©Èòµ: œÅÃÇ = D^(-1/2) Œ£ÃÇ D^(-1/2)
def compute_correlation_matrix(Sigma: np.ndarray) -> np.ndarray:
    """
    ‰ªéÂçèÊñπÂ∑ÆÁü©ÈòµËÆ°ÁÆóÁõ∏ÂÖ≥Áü©Èòµ
    
    Args:
        Sigma: ÂçèÊñπÂ∑ÆÁü©Èòµ
        
    Returns:
        Áõ∏ÂÖ≥Áü©Èòµ
    """
    D_inv_sqrt = np.diag(1 / np.sqrt(np.diag(Sigma)))
    return D_inv_sqrt @ Sigma @ D_inv_sqrt

rho_hat = compute_correlation_matrix(Sigma_hat)
print(f"\nÊ†∑Êú¨Áõ∏ÂÖ≥Áü©Èòµ œÅÃÇ:")
print(rho_hat)

# È™åËØÅ‰∏énumpyÂÜÖÁΩÆÂáΩÊï∞ÁöÑ‰∏ÄËá¥ÊÄß
Sigma_numpy = np.cov(returns_df.values.T)
rho_numpy = np.corrcoef(returns_df.values.T)

print(f"\n‚úÖ È™åËØÅ: ‰∏énumpy.covÁªìÊûú‰∏ÄËá¥: {np.allclose(Sigma_hat, Sigma_numpy)}")
print(f"‚úÖ È™åËØÅ: ‰∏énumpy.corrcoefÁªìÊûú‰∏ÄËá¥: {np.allclose(rho_hat, rho_numpy)}")
````

````python
# 4.3.3 ÂçèÊñπÂ∑ÆÊ§≠ÂúÜÂèØËßÜÂåñ
def plot_covariance_ellipse(returns: pd.DataFrame, asset1: str, asset2: str, 
                           sigma_levels: list = [1, 2]) -> None:
    """
    ÁªòÂà∂‰∏§‰∏™ËµÑ‰∫ßÁöÑÂçèÊñπÂ∑ÆÊ§≠ÂúÜ
    
    Args:
        returns: Êî∂ÁõäÊï∞ÊçÆDataFrame
        asset1, asset2: Ë¶ÅÁªòÂà∂ÁöÑ‰∏§‰∏™ËµÑ‰∫ßÂêçÁß∞
        sigma_levels: Ë¶ÅÁªòÂà∂ÁöÑœÉÊ∞¥Âπ≥ÂàóË°®
    """
    # ÊèêÂèñÊï∞ÊçÆ
    data = returns[[asset1, asset2]].values
    mu = data.mean(axis=0)
    Sigma_2d = np.cov(data.T)
    
    # ÁâπÂæÅÂàÜËß£
    eigenvals, eigenvecs = np.linalg.eigh(Sigma_2d)
    eigenvals = np.maximum(eigenvals, 1e-10)  # Á°Æ‰øùÈùûË¥ü
    
    # ÊåâÁâπÂæÅÂÄºÈôçÂ∫èÊéíÂàó
    idx = np.argsort(eigenvals)[::-1]
    eigenvals = eigenvals[idx]
    eigenvecs = eigenvecs[:, idx]
    
    # ÁªòÂà∂Êï£ÁÇπÂõæ
    plt.figure(figsize=(10, 8))
    plt.scatter(data[:, 0], data[:, 1], alpha=0.6, s=20, color='steelblue', 
               label='ËßÇÊµãÊï∞ÊçÆ')
    plt.scatter(mu[0], mu[1], color='red', s=100, marker='x', 
               label=f'ÂùáÂÄº ({mu[0]:.4f}, {mu[1]:.4f})')
    
    # ÁªòÂà∂Ê§≠ÂúÜ
    angles = np.linspace(0, 2*np.pi, 100)
    for sigma in sigma_levels:
        # Ê§≠ÂúÜÂèÇÊï∞
        a = sigma * np.sqrt(eigenvals[0])  # ÈïøËΩ¥
        b = sigma * np.sqrt(eigenvals[1])  # Áü≠ËΩ¥
        
        # Ê§≠ÂúÜÂùêÊ†á
        ellipse_x = a * np.cos(angles)
        ellipse_y = b * np.sin(angles)
        
        # ÊóãËΩ¨Ê§≠ÂúÜ
        rotation_matrix = eigenvecs
        ellipse_rotated = rotation_matrix @ np.array([ellipse_x, ellipse_y])
        ellipse_x_rot = ellipse_rotated[0] + mu[0]
        ellipse_y_rot = ellipse_rotated[1] + mu[1]
        
        plt.plot(ellipse_x_rot, ellipse_y_rot, '--', linewidth=2, 
                label=f'{œÉ}œÉ Ê§≠ÂúÜ')
    
    # ÁªòÂà∂‰∏ªÊñπÂêëÁÆ≠Â§¥
    for i, (eigenval, eigenvec) in enumerate(zip(eigenvals, eigenvecs.T)):
        length = 2 * np.sqrt(eigenval)
        plt.arrow(mu[0], mu[1], 
                 eigenvec[0] * length, eigenvec[1] * length,
                 head_width=0.01, head_length=0.01, fc='red', ec='red',
                 label=f'‰∏ªÊñπÂêë {i+1} (Œª={eigenval:.4f})')
    
    plt.xlabel(f'{asset1} Êî∂Áõä')
    plt.ylabel(f'{asset2} Êî∂Áõä')
    plt.title(f'[Day1] {asset1} vs {asset2} ÂçèÊñπÂ∑ÆÊ§≠ÂúÜ', fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.axis('equal')
    plt.tight_layout()
    plt.savefig(f'{OUTDIR}/covariance_ellipse_{asset1}_{asset2}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# ÁªòÂà∂Bond vs EquityÁöÑÂçèÊñπÂ∑ÆÊ§≠ÂúÜ
plot_covariance_ellipse(returns_df, 'Bond', 'Equity')

print("üìä ÂçèÊñπÂ∑ÆÊ§≠ÂúÜËß£Èáä:")
print("   - Ê§≠ÂúÜË∂äÊâÅÔºåËØ¥Êòé‰∏§‰∏™ËµÑ‰∫ßÁõ∏ÂÖ≥ÊÄßË∂äÂº∫")
print("   - Ê§≠ÂúÜË∂äÂúÜÔºåËØ¥Êòé‰∏§‰∏™ËµÑ‰∫ßÁõ∏ÂÖ≥ÊÄßË∂äÂº±")
print("   - ‰∏ªÊñπÂêëÁÆ≠Â§¥ÊåáÂêëÊúÄÂ§ßÊñπÂ∑ÆÊñπÂêë")
print("   - Bond‰∏éEquityÁöÑË¥üÁõ∏ÂÖ≥‰ΩìÁé∞Âú®Ê§≠ÂúÜÁöÑÂÄæÊñúÊñπÂêë‰∏ä")
````

````python
# 5.1 ÂçèÊñπÂ∑ÆÁü©ÈòµÁöÑÁâπÂæÅÂàÜËß£
print("üîç ÂçèÊñπÂ∑ÆÁü©ÈòµÁâπÂæÅÂàÜËß£")
print("=" * 50)

# ÂØπÊ†∑Êú¨ÂçèÊñπÂ∑ÆÁü©ÈòµËøõË°åÁâπÂæÅÂàÜËß£
eigenvals, eigenvecs = np.linalg.eigh(Sigma_hat)

# ÊåâÁâπÂæÅÂÄºÈôçÂ∫èÊéíÂàó
idx = np.argsort(eigenvals)[::-1]
eigenvals = eigenvals[idx]
eigenvecs = eigenvecs[:, idx]

print(f"ÁâπÂæÅÂÄº (ÊåâÈôçÂ∫è): {eigenvals}")
print(f"\nÁâπÂæÅÂêëÈáè (ÂàóÂêëÈáè):")
for i, (val, vec) in enumerate(zip(eigenvals, eigenvecs.T)):
    print(f"Œª{i+1} = {val:.6f}: {vec}")

# ËÆ°ÁÆóËß£ÈáäÊñπÂ∑ÆÊØî‰æã
explained_variance_ratio = eigenvals / np.sum(eigenvals)
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

print(f"\nËß£ÈáäÊñπÂ∑ÆÊØî‰æã:")
for i, (ratio, cum_ratio) in enumerate(zip(explained_variance_ratio, cumulative_variance_ratio)):
    print(f"PC{i+1}: {ratio:.1%} (Á¥ØÁßØ: {cum_ratio:.1%})")

# ÈáëËûçËß£Èáä
print(f"\nüí° ÈáëËûçËß£Èáä:")
print(f"   - ÊúÄÂ§ßÁâπÂæÅÂÄº Œª1 = {eigenvals[0]:.6f} ÂØπÂ∫î'Â∏ÇÂú∫È£éÈô©Âõ†Â≠ê'")
print(f"   - Ââç‰∏§‰∏™‰∏ªÊàêÂàÜËß£Èáä‰∫Ü {cumulative_variance_ratio[1]:.1%} ÁöÑÊÄªÈ£éÈô©")
print(f"   - ÁâπÂæÅÂêëÈáèÊòæÁ§∫ÂêÑËµÑ‰∫ßÂú®‰∏ªË¶ÅÈ£éÈô©Âõ†Â≠ê‰∏äÁöÑ'ËΩΩËç∑'")
````

````python
# 5.2 ÁâπÂæÅÂÄºÂèØËßÜÂåñ
print("\nüìä ÁâπÂæÅÂÄºÂèØËßÜÂåñ")
print("=" * 50)

# ÂàõÂª∫ÁâπÂæÅÂÄºÂàÜÊûêÂõæ
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 5.2.1 ÁâπÂæÅÂÄºÊù°ÂΩ¢Âõæ
bars = ax1.bar(range(1, len(eigenvals)+1), eigenvals, 
               color=['red', 'orange', 'green', 'blue'][:len(eigenvals)],
               alpha=0.7, edgecolor='black', linewidth=1)
ax1.set_xlabel('‰∏ªÊàêÂàÜ')
ax1.set_ylabel('ÁâπÂæÅÂÄº')
ax1.set_title('[Day1] ÂçèÊñπÂ∑ÆÁü©ÈòµÁâπÂæÅÂÄº', fontweight='bold')
ax1.grid(True, alpha=0.3)

# Ê∑ªÂä†Êï∞ÂÄºÊ†áÁ≠æ
for i, (bar, val) in enumerate(zip(bars, eigenvals)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'{val:.4f}', ha='center', va='bottom', fontweight='bold')

# 5.2.2 Á¥ØÁßØËß£ÈáäÊñπÂ∑ÆÊØî‰æã
ax2.plot(range(1, len(cumulative_variance_ratio)+1), cumulative_variance_ratio, 
         'o-', linewidth=2, markersize=8, color='darkblue')
ax2.set_xlabel('‰∏ªÊàêÂàÜÊï∞Èáè')
ax2.set_ylabel('Á¥ØÁßØËß£ÈáäÊñπÂ∑ÆÊØî‰æã')
ax2.set_title('[Day1] Á¥ØÁßØËß£ÈáäÊñπÂ∑ÆÊØî‰æã', fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_ylim(0, 1.1)

# Ê∑ªÂä†Êï∞ÂÄºÊ†áÁ≠æ
for i, (x, y) in enumerate(zip(range(1, len(cumulative_variance_ratio)+1), cumulative_variance_ratio)):
    ax2.text(x, y + 0.02, f'{y:.1%}', ha='center', va='bottom', fontweight='bold')

# Ê∑ªÂä†90%Âíå95%ÂèÇËÄÉÁ∫ø
ax2.axhline(y=0.9, color='red', linestyle='--', alpha=0.7, label='90%')
ax2.axhline(y=0.95, color='orange', linestyle='--', alpha=0.7, label='95%')
ax2.legend()

plt.tight_layout()
plt.savefig(f'{OUTDIR}/eigenvalue_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"üìà ‰∏ªÊàêÂàÜÂàÜÊûêÁªìÊûú:")
print(f"   - Á¨¨1‰∏ªÊàêÂàÜËß£Èáä {explained_variance_ratio[0]:.1%} ÁöÑÊñπÂ∑Æ")
print(f"   - Ââç2‰∏™‰∏ªÊàêÂàÜËß£Èáä {cumulative_variance_ratio[1]:.1%} ÁöÑÊñπÂ∑Æ")
print(f"   - ÈúÄË¶Å {np.argmax(cumulative_variance_ratio >= 0.9) + 1} ‰∏™‰∏ªÊàêÂàÜËææÂà∞90%Ëß£ÈáäÂ∫¶")
print(f"   - ÈúÄË¶Å {np.argmax(cumulative_variance_ratio >= 0.95) + 1} ‰∏™‰∏ªÊàêÂàÜËææÂà∞95%Ëß£ÈáäÂ∫¶")
````

````python
# 5.3 ‰∏ªÈ£éÈô©ÊñπÂêëÂèØËßÜÂåñ
print("\nüéØ ‰∏ªÈ£éÈô©ÊñπÂêëÂèØËßÜÂåñ")
print("=" * 50)

# ÈÄâÊã©BondÂíåEquityËøõË°å2DÂèØËßÜÂåñ
bond_idx = ASSETS.index('Bond')
equity_idx = ASSETS.index('Equity')

# ÊèêÂèñÊï∞ÊçÆ
bond_returns = returns_df['Bond'].values
equity_returns = returns_df['Equity'].values

# ËÆ°ÁÆó2DÂçèÊñπÂ∑ÆÁü©Èòµ
data_2d = np.column_stack([bond_returns, equity_returns])
mu_2d = data_2d.mean(axis=0)
Sigma_2d = np.cov(data_2d.T)

# 2DÁâπÂæÅÂàÜËß£
eigenvals_2d, eigenvecs_2d = np.linalg.eigh(Sigma_2d)
idx_2d = np.argsort(eigenvals_2d)[::-1]
eigenvals_2d = eigenvals_2d[idx_2d]
eigenvecs_2d = eigenvecs_2d[:, idx_2d]

# ÁªòÂà∂Êï£ÁÇπÂõæÂíå‰∏ªÊñπÂêë
plt.figure(figsize=(12, 10))
plt.scatter(bond_returns, equity_returns, alpha=0.6, s=20, color='steelblue', 
           label='ËßÇÊµãÊï∞ÊçÆ')

# ÁªòÂà∂ÂùáÂÄºÁÇπ
plt.scatter(mu_2d[0], mu_2d[1], color='red', s=100, marker='x', 
           label=f'ÂùáÂÄº ({mu_2d[0]:.4f}, {mu_2d[1]:.4f})')

# ÁªòÂà∂‰∏ªÊñπÂêëÁÆ≠Â§¥
colors = ['red', 'orange']
for i, (eigenval, eigenvec) in enumerate(zip(eigenvals_2d, eigenvecs_2d.T)):
    # ÁÆ≠Â§¥ÈïøÂ∫¶‰∏éÁâπÂæÅÂÄºÂπ≥ÊñπÊ†πÊàêÊØî‰æã
    length = 3 * np.sqrt(eigenval)
    plt.arrow(mu_2d[0], mu_2d[1], 
             eigenvec[0] * length, eigenvec[1] * length,
             head_width=0.005, head_length=0.005, 
             fc=colors[i], ec=colors[i], linewidth=3,
             label=f'‰∏ªÊñπÂêë {i+1} (Œª={eigenval:.6f})')

# ÁªòÂà∂1œÉÂíå2œÉÊ§≠ÂúÜ
angles = np.linspace(0, 2*np.pi, 100)
for sigma in [1, 2]:
    a = sigma * np.sqrt(eigenvals_2d[0])
    b = sigma * np.sqrt(eigenvals_2d[1])
    
    ellipse_x = a * np.cos(angles)
    ellipse_y = b * np.sin(angles)
    
    # ÊóãËΩ¨Ê§≠ÂúÜ
    ellipse_rotated = eigenvecs_2d @ np.array([ellipse_x, ellipse_y])
    ellipse_x_rot = ellipse_rotated[0] + mu_2d[0]
    ellipse_y_rot = ellipse_rotated[1] + mu_2d[1]
    
    plt.plot(ellipse_x_rot, ellipse_y_rot, '--', linewidth=2, alpha=0.7,
            label=f'{œÉ}œÉ Ê§≠ÂúÜ')

plt.xlabel('Bond Êî∂Áõä', fontsize=12)
plt.ylabel('Equity Êî∂Áõä', fontsize=12)
plt.title('[Day1] Bond vs Equity ‰∏ªÈ£éÈô©ÊñπÂêëÂàÜÊûê', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.tight_layout()
plt.savefig(f'{OUTDIR}/principal_directions.png', dpi=300, bbox_inches='tight')
plt.show()

# ÂàÜÊûê‰∏ªÊñπÂêë
print(f"üîç ‰∏ªÊñπÂêëÂàÜÊûê:")
print(f"   Á¨¨1‰∏ªÊñπÂêë (ÊúÄÂ§ßÈ£éÈô©): {eigenvecs_2d[:, 0]}")
print(f"   Á¨¨2‰∏ªÊñπÂêë (Ê¨°Â§ßÈ£éÈô©): {eigenvecs_2d[:, 1]}")
print(f"   Á¨¨1‰∏ªÊñπÂêëËß£Èáä {eigenvals_2d[0]/np.sum(eigenvals_2d):.1%} ÁöÑÊñπÂ∑Æ")
print(f"   Á¨¨2‰∏ªÊñπÂêëËß£Èáä {eigenvals_2d[1]/np.sum(eigenvals_2d):.1%} ÁöÑÊñπÂ∑Æ")

# ÈáëËûçËß£Èáä
print(f"\nüí° ÈáëËûçËß£Èáä:")
print(f"   - Á¨¨1‰∏ªÊñπÂêë: Â∏ÇÂú∫Á≥ªÁªüÊÄßÈ£éÈô©ÔºåBondÂíåEquityÂêåÂêëÂèòÂä®")
print(f"   - Á¨¨2‰∏ªÊñπÂêë: Áõ∏ÂØπÈ£éÈô©ÔºåBondÂíåEquityÂèçÂêëÂèòÂä®")
print(f"   - Ê§≠ÂúÜÁöÑÈïøËΩ¥ÊñπÂêëÂØπÂ∫îÊúÄÂ§ßÈ£éÈô©ÊñπÂêë")
print(f"   - Ê§≠ÂúÜÁöÑÁü≠ËΩ¥ÊñπÂêëÂØπÂ∫îÊúÄÂ∞èÈ£éÈô©ÊñπÂêë")
````

## Portfolio Variance & GradientÔΩúÁªÑÂêàÈ£éÈô©‰∏éÊ¢ØÂ∫¶

Áé∞Âú®Êàë‰ª¨Â∞ÜÂ≠¶‰π†Â¶Ç‰ΩïËÆ°ÁÆóÊäïËµÑÁªÑÂêàÁöÑÈ£éÈô©Ôºå‰ª•ÂèäÂ¶Ç‰ΩïÈÄöËøáÊ¢ØÂ∫¶‰∏ãÈôçÊù•‰ºòÂåñÊäïËµÑÁªÑÂêà„ÄÇ


````python
# 6.1 ÊäïËµÑÁªÑÂêàÁªüËÆ°ÂáΩÊï∞
print("üìä ÊäïËµÑÁªÑÂêàÁªüËÆ°ÂáΩÊï∞")
print("=" * 50)

def portfolio_stats(mu: np.ndarray, Sigma: np.ndarray, w: np.ndarray) -> dict:
    """
    ËÆ°ÁÆóÊäïËµÑÁªÑÂêàÁöÑÁªüËÆ°ÊåáÊ†á
    
    Args:
        mu: ÊúüÊúõÊî∂ÁõäÂêëÈáè (n,)
        Sigma: ÂçèÊñπÂ∑ÆÁü©Èòµ (n, n)
        w: ÊùÉÈáçÂêëÈáè (n,)
        
    Returns:
        ÂåÖÂê´ÁªÑÂêàÁªüËÆ°ÊåáÊ†áÁöÑÂ≠óÂÖ∏
    """
    # Á°Æ‰øùÊùÉÈáçÂíå‰∏∫1
    w = w / np.sum(w)
    
    # ÁªÑÂêàÊúüÊúõÊî∂Áõä: Œº_p = w^T * Œº
    portfolio_return = np.dot(w, mu)
    
    # ÁªÑÂêàÊñπÂ∑Æ: œÉ_p¬≤ = w^T * Œ£ * w
    portfolio_variance = np.dot(w, np.dot(Sigma, w))
    
    # ÁªÑÂêàÊ†áÂáÜÂ∑Æ
    portfolio_std = np.sqrt(portfolio_variance)
    
    # Â§èÊôÆÊØîÁéá (ÂÅáËÆæÊó†È£éÈô©Âà©Áéá‰∏∫0)
    sharpe_ratio = portfolio_return / portfolio_std if portfolio_std > 0 else 0
    
    # È£éÈô©Ë¥°ÁåÆ (ÊØè‰∏™ËµÑ‰∫ßÂØπÁªÑÂêàÊñπÂ∑ÆÁöÑË¥°ÁåÆ)
    risk_contributions = w * (Sigma @ w)
    
    return {
        'weights': w,
        'expected_return': portfolio_return,
        'variance': portfolio_variance,
        'std': portfolio_std,
        'sharpe_ratio': sharpe_ratio,
        'risk_contributions': risk_contributions
    }

def risk_gradient(Sigma: np.ndarray, w: np.ndarray) -> np.ndarray:
    """
    ËÆ°ÁÆóÁªÑÂêàÊñπÂ∑ÆÂØπÊùÉÈáçÁöÑÊ¢ØÂ∫¶
    
    Êï∞Â≠¶ÂÖ¨Âºè: ‚àá_w œÉ_p¬≤ = 2 * Œ£ * w
    
    Args:
        Sigma: ÂçèÊñπÂ∑ÆÁü©Èòµ (n, n)
        w: ÊùÉÈáçÂêëÈáè (n,)
        
    Returns:
        Ê¢ØÂ∫¶ÂêëÈáè (n,)
    """
    return 2 * Sigma @ w

# ÊµãËØïÂáΩÊï∞
print("‚úÖ ÊäïËµÑÁªÑÂêàÁªüËÆ°ÂáΩÊï∞ÂÆö‰πâÂÆåÊàê")
print("‚úÖ È£éÈô©Ê¢ØÂ∫¶ÂáΩÊï∞ÂÆö‰πâÂÆåÊàê")
````

````python
# 6.2 ÊäïËµÑÁªÑÂêàÂÆûÈ™å
print("\nüß™ ÊäïËµÑÁªÑÂêàÂÆûÈ™å")
print("=" * 50)

# ÂÆûÈ™å1: Á≠âÊùÉÈáçÊäïËµÑÁªÑÂêà
print("ÂÆûÈ™å1: Á≠âÊùÉÈáçÊäïËµÑÁªÑÂêà")
w_equal = np.ones(n_assets) / n_assets
stats_equal = portfolio_stats(mu_hat, Sigma_hat, w_equal)

print(f"ÊùÉÈáç: {stats_equal['weights']}")
print(f"ÊúüÊúõÊî∂Áõä: {stats_equal['expected_return']:.6f}")
print(f"Ê†áÂáÜÂ∑Æ: {stats_equal['std']:.6f}")
print(f"Â§èÊôÆÊØîÁéá: {stats_equal['sharpe_ratio']:.6f}")
print(f"È£éÈô©Ë¥°ÁåÆ: {stats_equal['risk_contributions']}")

# ÂÆûÈ™å2: ÂÖ®ËÇ°Á•®ÊäïËµÑÁªÑÂêà
print(f"\nÂÆûÈ™å2: ÂÖ®ËÇ°Á•®ÊäïËµÑÁªÑÂêà")
w_stock = np.zeros(n_assets)
w_stock[ASSETS.index('Equity')] = 1.0
stats_stock = portfolio_stats(mu_hat, Sigma_hat, w_stock)

print(f"ÊùÉÈáç: {stats_stock['weights']}")
print(f"ÊúüÊúõÊî∂Áõä: {stats_stock['expected_return']:.6f}")
print(f"Ê†áÂáÜÂ∑Æ: {stats_stock['std']:.6f}")
print(f"Â§èÊôÆÊØîÁéá: {stats_stock['sharpe_ratio']:.6f}")

# ÂÆûÈ™å3: È£éÈô©Ê¢ØÂ∫¶‰∏ãÈôç
print(f"\nÂÆûÈ™å3: È£éÈô©Ê¢ØÂ∫¶‰∏ãÈôç")
w_test = np.array([0.1, 0.2, 0.6, 0.1])  # ÊµãËØïÊùÉÈáç
w_test = w_test / np.sum(w_test)  # ÂΩí‰∏ÄÂåñ

print(f"ÂàùÂßãÊùÉÈáç: {w_test}")
print(f"ÂàùÂßãÊñπÂ∑Æ: {portfolio_stats(mu_hat, Sigma_hat, w_test)['variance']:.6f}")

# ËÆ°ÁÆóÊ¢ØÂ∫¶
gradient = risk_gradient(Sigma_hat, w_test)
print(f"È£éÈô©Ê¢ØÂ∫¶: {gradient}")

# Ê≤øË¥üÊ¢ØÂ∫¶ÊñπÂêëÊõ¥Êñ∞ÊùÉÈáç
learning_rate = 0.01
w_new = w_test - learning_rate * gradient
w_new = w_new / np.sum(w_new)  # ÈáçÊñ∞ÂΩí‰∏ÄÂåñ

print(f"Êõ¥Êñ∞ÂêéÊùÉÈáç: {w_new}")
print(f"Êõ¥Êñ∞ÂêéÊñπÂ∑Æ: {portfolio_stats(mu_hat, Sigma_hat, w_new)['variance']:.6f}")

# È™åËØÅÊ¢ØÂ∫¶‰∏ãÈôçÊòØÂê¶Èôç‰Ωé‰∫ÜÊñπÂ∑Æ
old_var = portfolio_stats(mu_hat, Sigma_hat, w_test)['variance']
new_var = portfolio_stats(mu_hat, Sigma_hat, w_new)['variance']
print(f"ÊñπÂ∑ÆÂèòÂåñ: {old_var - new_var:.6f} ({'‚úÖ Èôç‰Ωé' if new_var < old_var else '‚ùå Â¢ûÂä†'})")
````

````python
# 6.3 È£éÈô©Ë¥°ÁåÆÂèØËßÜÂåñ
print("\nüìä È£éÈô©Ë¥°ÁåÆÂèØËßÜÂåñ")
print("=" * 50)

# ËÆ°ÁÆó‰∏çÂêåÊäïËµÑÁªÑÂêàÁöÑÈ£éÈô©Ë¥°ÁåÆ
portfolios = {
    'Á≠âÊùÉÈáç': w_equal,
    'ÂÖ®ËÇ°Á•®': w_stock,
    'ÊµãËØïÁªÑÂêà': w_test
}

# ÂàõÂª∫È£éÈô©Ë¥°ÁåÆÂØπÊØîÂõæ
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('[Day1] ‰∏çÂêåÊäïËµÑÁªÑÂêàÁöÑÈ£éÈô©Ë¥°ÁåÆÂàÜÊûê', fontsize=16, fontweight='bold')

for i, (name, weights) in enumerate(portfolios.items()):
    stats = portfolio_stats(mu_hat, Sigma_hat, weights)
    
    # ÁªòÂà∂È£éÈô©Ë¥°ÁåÆÊù°ÂΩ¢Âõæ
    bars = axes[i].bar(ASSETS, stats['risk_contributions'], 
                      color=['red', 'orange', 'green', 'blue'], alpha=0.7,
                      edgecolor='black', linewidth=1)
    
    axes[i].set_title(f'{name}ÁªÑÂêà', fontweight='bold')
    axes[i].set_ylabel('È£éÈô©Ë¥°ÁåÆ')
    axes[i].set_xlabel('ËµÑ‰∫ßÁ±ªÂà´')
    axes[i].grid(True, alpha=0.3)
    
    # Ê∑ªÂä†Êï∞ÂÄºÊ†áÁ≠æ
    for bar, val in zip(bars, stats['risk_contributions']):
        height = bar.get_height()
        axes[i].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                    f'{val:.4f}', ha='center', va='bottom', fontweight='bold')
    
    # Ê∑ªÂä†ÁªÑÂêàÁªüËÆ°‰ø°ÊÅØ
    axes[i].text(0.02, 0.98, f'ÊúüÊúõÊî∂Áõä: {stats["expected_return"]:.4f}\nÊ†áÂáÜÂ∑Æ: {stats["std"]:.4f}\nÂ§èÊôÆÊØîÁéá: {stats["sharpe_ratio"]:.4f}',
                transform=axes[i].transAxes, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.savefig(f'{OUTDIR}/risk_contributions.png', dpi=300, bbox_inches='tight')
plt.show()

# ÂàÜÊï£ÂåñÊïàÊûúÂàÜÊûê
print(f"üìà ÂàÜÊï£ÂåñÊïàÊûúÂàÜÊûê:")
print(f"   Á≠âÊùÉÈáçÁªÑÂêàÊ†áÂáÜÂ∑Æ: {stats_equal['std']:.6f}")
print(f"   ÂÖ®ËÇ°Á•®ÁªÑÂêàÊ†áÂáÜÂ∑Æ: {stats_stock['std']:.6f}")
print(f"   ÂàÜÊï£ÂåñÊî∂Áõä: {stats_stock['std'] - stats_equal['std']:.6f}")
print(f"   È£éÈô©Èôç‰ΩéÊØî‰æã: {(stats_stock['std'] - stats_equal['std']) / stats_stock['std']:.1%}")

print(f"\nüí° ÂÖ≥ÈîÆÊ¥ûÂØü:")
print(f"   - Á≠âÊùÉÈáçÁªÑÂêàÈÄöËøáÂàÜÊï£ÂåñÈôç‰Ωé‰∫ÜÈ£éÈô©")
print(f"   - È£éÈô©Ë¥°ÁåÆÊòæÁ§∫ÂêÑËµÑ‰∫ßÂØπÁªÑÂêàÈ£éÈô©ÁöÑÁõ∏ÂØπË¥°ÁåÆ")
print(f"   - Ë¥üÁõ∏ÂÖ≥ËµÑ‰∫ß(Â¶ÇBondÂíåEquity)ÊúâÂä©‰∫éÈôç‰ΩéÁªÑÂêàÈ£éÈô©")
print(f"   - Ê¢ØÂ∫¶‰∏ãÈôçÂèØ‰ª•Áî®‰∫éÈ£éÈô©‰ºòÂåñ")
````

## Finance MappingÔΩúÈáëËûçËØ≠‰πâÊò†Â∞Ñ

Áé∞Âú®ËÆ©Êàë‰ª¨Â∞ÜÊï∞Â≠¶Ê¶ÇÂøµ‰∏éÈáëËûçÂÆûË∑µËÅîÁ≥ªËµ∑Êù•ÔºåÁêÜËß£Ëøô‰∫õÊäΩË±°Ê¶ÇÂøµÂú®ÂÆûÈôÖÊäïËµÑ‰∏≠ÁöÑÊÑè‰πâ„ÄÇ
