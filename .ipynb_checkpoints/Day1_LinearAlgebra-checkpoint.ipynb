### 为什么线性代数是资产配置的"母语"？

在量化金融中，线性代数不仅仅是数学工具，更是我们理解金融市场的"语言"：

- **向量 = 收益截面**：每个资产在特定时期的收益可以表示为向量中的一个元素
- **矩阵 = 关系结构**：协方差矩阵捕捉了资产间的风险关联性
- **内积 = 协动关系**：两个资产收益的内积反映了它们的协动程度
- **特征分解 = 风险主方向**：协方差矩阵的特征向量揭示了市场的主要风险因子方向

## Setup｜环境与工具


````python
# 导入核心库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import linalg
import os
from pathlib import Path

# 设置随机种子确保结果可复现
rng = np.random.default_rng(42)

# 配置绘图风格
if seaborn_available:
    import seaborn as sns
    sns.set_style("whitegrid")
    sns.set_palette("husl")
    print("📊 使用 seaborn 专业绘图风格")
else:
    plt.style.use('default')
    print("📊 使用 matplotlib 默认绘图风格")

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# 定义常量
ASSETS = ["CashProxy", "Bond", "Equity", "Gold"]
n_assets = len(ASSETS)

# 年化参数
mu_annual = np.array([0.02, 0.045, 0.08, 0.03])        # 年化期望收益
sigma_annual = np.array([0.005, 0.06, 0.18, 0.12])     # 年化波动

# 相关系数矩阵（金融直觉：债券与股票负相关，黄金与股票低相关）
rho = np.array([
    [1.00,  0.10, -0.05,  0.00],  # CashProxy
    [0.10,  1.00, -0.30,  0.10],  # Bond  
    [-0.05,-0.30,  1.00,  0.15],  # Equity
    [0.00,  0.10,  0.15,  1.00],  # Gold
])

# 时间参数
TRADING_DAYS = 252  # 年交易日数
T = 750             # 样本长度（约3年）
OUTDIR = "./artifacts"

# 创建输出目录
Path(OUTDIR).mkdir(exist_ok=True)
print(f"📁 输出目录: {OUTDIR}")

print("🔧 环境配置完成！")
print(f"📊 资产类别: {ASSETS}")
print(f"📈 样本长度: {T} 个交易日")
print(f"🎯 年化期望收益: {mu_annual}")
print(f"📉 年化波动率: {sigma_annual}")
````

## Data Simulation｜模拟多资产收益

现在我们将使用预设的相关结构来构造协方差矩阵，并生成多资产的模拟收益数据。


````python
# 步骤1: 将年化参数转换为日度参数
mu_daily = mu_annual / TRADING_DAYS
sigma_daily = sigma_annual / np.sqrt(TRADING_DAYS)

print("📊 日度参数转换:")
print(f"   期望收益: {mu_daily}")
print(f"   波动率: {sigma_daily}")

# 步骤2: 构造协方差矩阵 Σ = D * ρ * D
# 其中 D = diag(σ) 是标准差对角矩阵
D = np.diag(sigma_daily)
Sigma = D @ rho @ D

print(f"\n📈 协方差矩阵 Σ (日度):")
print(Sigma)

# 步骤3: 数值稳定性检查与修正
def ensure_positive_definite(Sigma: np.ndarray, tol: float = 1e-8) -> np.ndarray:
    """
    确保协方差矩阵正定
    
    Args:
        Sigma: 协方差矩阵
        tol: 对角线扰动容忍度
        
    Returns:
        修正后的正定协方差矩阵
    """
    # 对称化
    Sigma_sym = (Sigma + Sigma.T) / 2
    
    # 检查正定性
    try:
        np.linalg.cholesky(Sigma_sym)
        print("✅ 协方差矩阵已正定")
        return Sigma_sym
    except np.linalg.LinAlgError:
        print("⚠️  协方差矩阵非正定，添加对角线扰动...")
        # 添加对角线扰动确保正定
        eigenvals = np.linalg.eigvals(Sigma_sym)
        min_eigenval = np.min(eigenvals)
        if min_eigenval <= 0:
            Sigma_corrected = Sigma_sym + (tol - min_eigenval) * np.eye(Sigma_sym.shape[0])
            print(f"   添加对角线扰动: {tol - min_eigenval:.2e}")
            return Sigma_corrected
        return Sigma_sym

Sigma = ensure_positive_definite(Sigma)

# 步骤4: Cholesky分解生成模拟数据
try:
    L = np.linalg.cholesky(Sigma)
    print("✅ Cholesky分解成功")
    
    # 生成标准正态随机数
    Z = rng.standard_normal((T, n_assets))
    
    # 应用Cholesky变换: R = μ + L * Z
    returns = mu_daily + Z @ L.T
    
    print(f"📊 生成 {T} 个日度收益样本")
    
except np.linalg.LinAlgError as e:
    print(f"❌ Cholesky分解失败: {e}")
    # 使用特征分解作为备选方案
    eigenvals, eigenvecs = np.linalg.eigh(Sigma)
    eigenvals = np.maximum(eigenvals, 1e-8)  # 确保非负
    L = eigenvecs @ np.diag(np.sqrt(eigenvals))
    Z = rng.standard_normal((T, n_assets))
    returns = mu_daily + Z @ L.T
    print("✅ 使用特征分解备选方案生成数据")

# 步骤5: 创建DataFrame并保存
returns_df = pd.DataFrame(returns, columns=ASSETS)
returns_df.index.name = 'Date'

# 保存为parquet格式
returns_df.to_parquet(f"{OUTDIR}/day1_returns.parquet")
print(f"💾 数据已保存至: {OUTDIR}/day1_returns.parquet")

# 显示数据摘要
print(f"\n📈 模拟数据摘要:")
print(f"   形状: {returns_df.shape}")
print(f"   均值: {returns_df.mean().values}")
print(f"   标准差: {returns_df.std().values}")
print(f"\n前5行数据:")
print(returns_df.head())
````

## Math Warm-up｜线性代数与统计最小全家桶

让我们回顾线性代数和统计学中的核心概念，这些是理解资产配置数学基础的关键。


````python
# 4.1 向量与矩阵基础
print("🔢 向量与矩阵基础")
print("=" * 50)

# 从数据中提取向量示例
sample_returns = returns_df.iloc[0].values  # 第一天的收益向量
print(f"收益向量示例 (第一天): {sample_returns}")
print(f"向量形状: {sample_returns.shape}")
print(f"向量转置: {sample_returns.T}")
print(f"L2范数: {np.linalg.norm(sample_returns):.6f}")

# 矩阵操作
print(f"\n矩阵形状: {returns_df.shape}")
print(f"矩阵转置形状: {returns_df.T.shape}")

# 内积与范数
def vector_operations_demo(x: np.ndarray, y: np.ndarray) -> None:
    """演示向量运算"""
    print(f"向量 x: {x}")
    print(f"向量 y: {y}")
    print(f"内积 x·y: {np.dot(x, y):.6f}")
    print(f"L2范数 ||x||: {np.linalg.norm(x):.6f}")
    print(f"单位向量 x/||x||: {x / np.linalg.norm(x)}")

# 使用前两天的数据演示
x = returns_df.iloc[0].values
y = returns_df.iloc[1].values
vector_operations_demo(x, y)
````

````python
# 4.2 统计量计算
print("\n📊 统计量计算")
print("=" * 50)

# 样本均值向量: μ̂ = (1/T) Σ r_t
mu_hat = returns_df.mean().values
print(f"样本均值向量 μ̂: {mu_hat}")

# 样本协方差矩阵: Σ̂ = (1/(T-1)) (R - μ̂)(R - μ̂)^T
def compute_sample_covariance(returns: np.ndarray) -> np.ndarray:
    """
    计算样本协方差矩阵
    
    Args:
        returns: T×n 收益矩阵
        
    Returns:
        n×n 协方差矩阵
    """
    T, n = returns.shape
    mu = returns.mean(axis=0)
    centered = returns - mu  # 去均值
    return (centered.T @ centered) / (T - 1)

Sigma_hat = compute_sample_covariance(returns_df.values)
print(f"\n样本协方差矩阵 Σ̂:")
print(Sigma_hat)

# 相关矩阵: ρ̂ = D^(-1/2) Σ̂ D^(-1/2)
def compute_correlation_matrix(Sigma: np.ndarray) -> np.ndarray:
    """
    从协方差矩阵计算相关矩阵
    
    Args:
        Sigma: 协方差矩阵
        
    Returns:
        相关矩阵
    """
    D_inv_sqrt = np.diag(1 / np.sqrt(np.diag(Sigma)))
    return D_inv_sqrt @ Sigma @ D_inv_sqrt

rho_hat = compute_correlation_matrix(Sigma_hat)
print(f"\n样本相关矩阵 ρ̂:")
print(rho_hat)

# 验证与numpy内置函数的一致性
Sigma_numpy = np.cov(returns_df.values.T)
rho_numpy = np.corrcoef(returns_df.values.T)

print(f"\n✅ 验证: 与numpy.cov结果一致: {np.allclose(Sigma_hat, Sigma_numpy)}")
print(f"✅ 验证: 与numpy.corrcoef结果一致: {np.allclose(rho_hat, rho_numpy)}")
````

````python
# 4.3 可视化分析
print("\n📈 可视化分析")
print("=" * 50)

# 创建图形
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('[Day1] 多资产收益分布与相关性分析', fontsize=16, fontweight='bold')

# 4.3.1 收益分布直方图
for i, asset in enumerate(ASSETS):
    row, col = i // 2, i % 2
    axes[row, col].hist(returns_df[asset], bins=30, alpha=0.7, density=True, 
                       color=f'C{i}', edgecolor='black', linewidth=0.5)
    axes[row, col].set_title(f'{asset} 收益分布', fontweight='bold')
    axes[row, col].set_xlabel('日度收益')
    axes[row, col].set_ylabel('密度')
    axes[row, col].grid(True, alpha=0.3)
    
    # 添加统计信息
    mean_val = returns_df[asset].mean()
    std_val = returns_df[asset].std()
    axes[row, col].axvline(mean_val, color='red', linestyle='--', 
                          label=f'均值: {mean_val:.4f}')
    axes[row, col].axvline(mean_val + std_val, color='orange', linestyle=':', 
                          label=f'±1σ: {std_val:.4f}')
    axes[row, col].axvline(mean_val - std_val, color='orange', linestyle=':')
    axes[row, col].legend()

plt.tight_layout()
plt.savefig(f'{OUTDIR}/returns_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# 4.3.2 相关矩阵热力图
plt.figure(figsize=(10, 8))

if seaborn_available:
    # 使用seaborn绘制更美观的热力图
    mask = np.triu(np.ones_like(rho_hat, dtype=bool))  # 只显示下三角
    sns.heatmap(rho_hat, mask=mask, annot=True, cmap='RdBu_r', center=0,
                square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
                xticklabels=ASSETS, yticklabels=ASSETS)
else:
    # 使用matplotlib绘制热力图
    im = plt.imshow(rho_hat, cmap='RdBu_r', aspect='equal', vmin=-1, vmax=1)
    plt.colorbar(im, shrink=0.8)
    
    # 添加数值标注
    for i in range(len(ASSETS)):
        for j in range(len(ASSETS)):
            plt.text(j, i, f'{rho_hat[i, j]:.2f}', 
                    ha='center', va='center', fontweight='bold')
    
    plt.xticks(range(len(ASSETS)), ASSETS)
    plt.yticks(range(len(ASSETS)), ASSETS)

plt.title('[Day1] 资产相关性热力图', fontsize=14, fontweight='bold')
plt.xlabel('资产类别')
plt.ylabel('资产类别')
plt.tight_layout()
plt.savefig(f'{OUTDIR}/correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
````

````python
# 4.3.3 协方差椭圆可视化
def plot_covariance_ellipse(returns: pd.DataFrame, asset1: str, asset2: str, 
                           sigma_levels: list = [1, 2]) -> None:
    """
    绘制两个资产的协方差椭圆
    
    Args:
        returns: 收益数据DataFrame
        asset1, asset2: 要绘制的两个资产名称
        sigma_levels: 要绘制的σ水平列表
    """
    # 提取数据
    data = returns[[asset1, asset2]].values
    mu = data.mean(axis=0)
    Sigma_2d = np.cov(data.T)
    
    # 特征分解
    eigenvals, eigenvecs = np.linalg.eigh(Sigma_2d)
    eigenvals = np.maximum(eigenvals, 1e-10)  # 确保非负
    
    # 按特征值降序排列
    idx = np.argsort(eigenvals)[::-1]
    eigenvals = eigenvals[idx]
    eigenvecs = eigenvecs[:, idx]
    
    # 绘制散点图
    plt.figure(figsize=(10, 8))
    plt.scatter(data[:, 0], data[:, 1], alpha=0.6, s=20, color='steelblue', 
               label='观测数据')
    plt.scatter(mu[0], mu[1], color='red', s=100, marker='x', 
               label=f'均值 ({mu[0]:.4f}, {mu[1]:.4f})')
    
    # 绘制椭圆
    angles = np.linspace(0, 2*np.pi, 100)
    for sigma in sigma_levels:
        # 椭圆参数
        a = sigma * np.sqrt(eigenvals[0])  # 长轴
        b = sigma * np.sqrt(eigenvals[1])  # 短轴
        
        # 椭圆坐标
        ellipse_x = a * np.cos(angles)
        ellipse_y = b * np.sin(angles)
        
        # 旋转椭圆
        rotation_matrix = eigenvecs
        ellipse_rotated = rotation_matrix @ np.array([ellipse_x, ellipse_y])
        ellipse_x_rot = ellipse_rotated[0] + mu[0]
        ellipse_y_rot = ellipse_rotated[1] + mu[1]
        
        plt.plot(ellipse_x_rot, ellipse_y_rot, '--', linewidth=2, 
                label=f'{σ}σ 椭圆')
    
    # 绘制主方向箭头
    for i, (eigenval, eigenvec) in enumerate(zip(eigenvals, eigenvecs.T)):
        length = 2 * np.sqrt(eigenval)
        plt.arrow(mu[0], mu[1], 
                 eigenvec[0] * length, eigenvec[1] * length,
                 head_width=0.01, head_length=0.01, fc='red', ec='red',
                 label=f'主方向 {i+1} (λ={eigenval:.4f})')
    
    plt.xlabel(f'{asset1} 收益')
    plt.ylabel(f'{asset2} 收益')
    plt.title(f'[Day1] {asset1} vs {asset2} 协方差椭圆', fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.axis('equal')
    plt.tight_layout()
    plt.savefig(f'{OUTDIR}/covariance_ellipse_{asset1}_{asset2}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 绘制Bond vs Equity的协方差椭圆
plot_covariance_ellipse(returns_df, 'Bond', 'Equity')

print("📊 协方差椭圆解释:")
print("   - 椭圆越扁，说明两个资产相关性越强")
print("   - 椭圆越圆，说明两个资产相关性越弱")
print("   - 主方向箭头指向最大方差方向")
print("   - Bond与Equity的负相关体现在椭圆的倾斜方向上")
````

## Eigen & PCA Intuition｜特征分解与主风险方向

协方差矩阵的特征分解揭示了市场的主要风险因子方向。让我们深入理解这一重要概念。


````python
# 5.1 协方差矩阵的特征分解
print("🔍 协方差矩阵特征分解")
print("=" * 50)

# 对样本协方差矩阵进行特征分解
eigenvals, eigenvecs = np.linalg.eigh(Sigma_hat)

# 按特征值降序排列
idx = np.argsort(eigenvals)[::-1]
eigenvals = eigenvals[idx]
eigenvecs = eigenvecs[:, idx]

print(f"特征值 (按降序): {eigenvals}")
print(f"\n特征向量 (列向量):")
for i, (val, vec) in enumerate(zip(eigenvals, eigenvecs.T)):
    print(f"λ{i+1} = {val:.6f}: {vec}")

# 计算解释方差比例
explained_variance_ratio = eigenvals / np.sum(eigenvals)
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

print(f"\n解释方差比例:")
for i, (ratio, cum_ratio) in enumerate(zip(explained_variance_ratio, cumulative_variance_ratio)):
    print(f"PC{i+1}: {ratio:.1%} (累积: {cum_ratio:.1%})")

# 金融解释
print(f"\n💡 金融解释:")
print(f"   - 最大特征值 λ1 = {eigenvals[0]:.6f} 对应'市场风险因子'")
print(f"   - 前两个主成分解释了 {cumulative_variance_ratio[1]:.1%} 的总风险")
print(f"   - 特征向量显示各资产在主要风险因子上的'载荷'")
````

````python
# 5.2 特征值可视化
print("\n📊 特征值可视化")
print("=" * 50)

# 创建特征值分析图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 5.2.1 特征值条形图
bars = ax1.bar(range(1, len(eigenvals)+1), eigenvals, 
               color=['red', 'orange', 'green', 'blue'][:len(eigenvals)],
               alpha=0.7, edgecolor='black', linewidth=1)
ax1.set_xlabel('主成分')
ax1.set_ylabel('特征值')
ax1.set_title('[Day1] 协方差矩阵特征值', fontweight='bold')
ax1.grid(True, alpha=0.3)

# 添加数值标签
for i, (bar, val) in enumerate(zip(bars, eigenvals)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'{val:.4f}', ha='center', va='bottom', fontweight='bold')

# 5.2.2 累积解释方差比例
ax2.plot(range(1, len(cumulative_variance_ratio)+1), cumulative_variance_ratio, 
         'o-', linewidth=2, markersize=8, color='darkblue')
ax2.set_xlabel('主成分数量')
ax2.set_ylabel('累积解释方差比例')
ax2.set_title('[Day1] 累积解释方差比例', fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_ylim(0, 1.1)

# 添加数值标签
for i, (x, y) in enumerate(zip(range(1, len(cumulative_variance_ratio)+1), cumulative_variance_ratio)):
    ax2.text(x, y + 0.02, f'{y:.1%}', ha='center', va='bottom', fontweight='bold')

# 添加90%和95%参考线
ax2.axhline(y=0.9, color='red', linestyle='--', alpha=0.7, label='90%')
ax2.axhline(y=0.95, color='orange', linestyle='--', alpha=0.7, label='95%')
ax2.legend()

plt.tight_layout()
plt.savefig(f'{OUTDIR}/eigenvalue_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"📈 主成分分析结果:")
print(f"   - 第1主成分解释 {explained_variance_ratio[0]:.1%} 的方差")
print(f"   - 前2个主成分解释 {cumulative_variance_ratio[1]:.1%} 的方差")
print(f"   - 需要 {np.argmax(cumulative_variance_ratio >= 0.9) + 1} 个主成分达到90%解释度")
print(f"   - 需要 {np.argmax(cumulative_variance_ratio >= 0.95) + 1} 个主成分达到95%解释度")
````

````python
# 5.3 主风险方向可视化
print("\n🎯 主风险方向可视化")
print("=" * 50)

# 选择Bond和Equity进行2D可视化
bond_idx = ASSETS.index('Bond')
equity_idx = ASSETS.index('Equity')

# 提取数据
bond_returns = returns_df['Bond'].values
equity_returns = returns_df['Equity'].values

# 计算2D协方差矩阵
data_2d = np.column_stack([bond_returns, equity_returns])
mu_2d = data_2d.mean(axis=0)
Sigma_2d = np.cov(data_2d.T)

# 2D特征分解
eigenvals_2d, eigenvecs_2d = np.linalg.eigh(Sigma_2d)
idx_2d = np.argsort(eigenvals_2d)[::-1]
eigenvals_2d = eigenvals_2d[idx_2d]
eigenvecs_2d = eigenvecs_2d[:, idx_2d]

# 绘制散点图和主方向
plt.figure(figsize=(12, 10))
plt.scatter(bond_returns, equity_returns, alpha=0.6, s=20, color='steelblue', 
           label='观测数据')

# 绘制均值点
plt.scatter(mu_2d[0], mu_2d[1], color='red', s=100, marker='x', 
           label=f'均值 ({mu_2d[0]:.4f}, {mu_2d[1]:.4f})')

# 绘制主方向箭头
colors = ['red', 'orange']
for i, (eigenval, eigenvec) in enumerate(zip(eigenvals_2d, eigenvecs_2d.T)):
    # 箭头长度与特征值平方根成比例
    length = 3 * np.sqrt(eigenval)
    plt.arrow(mu_2d[0], mu_2d[1], 
             eigenvec[0] * length, eigenvec[1] * length,
             head_width=0.005, head_length=0.005, 
             fc=colors[i], ec=colors[i], linewidth=3,
             label=f'主方向 {i+1} (λ={eigenval:.6f})')

# 绘制1σ和2σ椭圆
angles = np.linspace(0, 2*np.pi, 100)
for sigma in [1, 2]:
    a = sigma * np.sqrt(eigenvals_2d[0])
    b = sigma * np.sqrt(eigenvals_2d[1])
    
    ellipse_x = a * np.cos(angles)
    ellipse_y = b * np.sin(angles)
    
    # 旋转椭圆
    ellipse_rotated = eigenvecs_2d @ np.array([ellipse_x, ellipse_y])
    ellipse_x_rot = ellipse_rotated[0] + mu_2d[0]
    ellipse_y_rot = ellipse_rotated[1] + mu_2d[1]
    
    plt.plot(ellipse_x_rot, ellipse_y_rot, '--', linewidth=2, alpha=0.7,
            label=f'{σ}σ 椭圆')

plt.xlabel('Bond 收益', fontsize=12)
plt.ylabel('Equity 收益', fontsize=12)
plt.title('[Day1] Bond vs Equity 主风险方向分析', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.tight_layout()
plt.savefig(f'{OUTDIR}/principal_directions.png', dpi=300, bbox_inches='tight')
plt.show()

# 分析主方向
print(f"🔍 主方向分析:")
print(f"   第1主方向 (最大风险): {eigenvecs_2d[:, 0]}")
print(f"   第2主方向 (次大风险): {eigenvecs_2d[:, 1]}")
print(f"   第1主方向解释 {eigenvals_2d[0]/np.sum(eigenvals_2d):.1%} 的方差")
print(f"   第2主方向解释 {eigenvals_2d[1]/np.sum(eigenvals_2d):.1%} 的方差")

# 金融解释
print(f"\n💡 金融解释:")
print(f"   - 第1主方向: 市场系统性风险，Bond和Equity同向变动")
print(f"   - 第2主方向: 相对风险，Bond和Equity反向变动")
print(f"   - 椭圆的长轴方向对应最大风险方向")
print(f"   - 椭圆的短轴方向对应最小风险方向")
````

## Portfolio Variance & Gradient｜组合风险与梯度

现在我们将学习如何计算投资组合的风险，以及如何通过梯度下降来优化投资组合。


````python
# 6.1 投资组合统计函数
print("📊 投资组合统计函数")
print("=" * 50)

def portfolio_stats(mu: np.ndarray, Sigma: np.ndarray, w: np.ndarray) -> dict:
    """
    计算投资组合的统计指标
    
    Args:
        mu: 期望收益向量 (n,)
        Sigma: 协方差矩阵 (n, n)
        w: 权重向量 (n,)
        
    Returns:
        包含组合统计指标的字典
    """
    # 确保权重和为1
    w = w / np.sum(w)
    
    # 组合期望收益: μ_p = w^T * μ
    portfolio_return = np.dot(w, mu)
    
    # 组合方差: σ_p² = w^T * Σ * w
    portfolio_variance = np.dot(w, np.dot(Sigma, w))
    
    # 组合标准差
    portfolio_std = np.sqrt(portfolio_variance)
    
    # 夏普比率 (假设无风险利率为0)
    sharpe_ratio = portfolio_return / portfolio_std if portfolio_std > 0 else 0
    
    # 风险贡献 (每个资产对组合方差的贡献)
    risk_contributions = w * (Sigma @ w)
    
    return {
        'weights': w,
        'expected_return': portfolio_return,
        'variance': portfolio_variance,
        'std': portfolio_std,
        'sharpe_ratio': sharpe_ratio,
        'risk_contributions': risk_contributions
    }

def risk_gradient(Sigma: np.ndarray, w: np.ndarray) -> np.ndarray:
    """
    计算组合方差对权重的梯度
    
    数学公式: ∇_w σ_p² = 2 * Σ * w
    
    Args:
        Sigma: 协方差矩阵 (n, n)
        w: 权重向量 (n,)
        
    Returns:
        梯度向量 (n,)
    """
    return 2 * Sigma @ w

# 测试函数
print("✅ 投资组合统计函数定义完成")
print("✅ 风险梯度函数定义完成")
````

````python
# 6.2 投资组合实验
print("\n🧪 投资组合实验")
print("=" * 50)

# 实验1: 等权重投资组合
print("实验1: 等权重投资组合")
w_equal = np.ones(n_assets) / n_assets
stats_equal = portfolio_stats(mu_hat, Sigma_hat, w_equal)

print(f"权重: {stats_equal['weights']}")
print(f"期望收益: {stats_equal['expected_return']:.6f}")
print(f"标准差: {stats_equal['std']:.6f}")
print(f"夏普比率: {stats_equal['sharpe_ratio']:.6f}")
print(f"风险贡献: {stats_equal['risk_contributions']}")

# 实验2: 全股票投资组合
print(f"\n实验2: 全股票投资组合")
w_stock = np.zeros(n_assets)
w_stock[ASSETS.index('Equity')] = 1.0
stats_stock = portfolio_stats(mu_hat, Sigma_hat, w_stock)

print(f"权重: {stats_stock['weights']}")
print(f"期望收益: {stats_stock['expected_return']:.6f}")
print(f"标准差: {stats_stock['std']:.6f}")
print(f"夏普比率: {stats_stock['sharpe_ratio']:.6f}")

# 实验3: 风险梯度下降
print(f"\n实验3: 风险梯度下降")
w_test = np.array([0.1, 0.2, 0.6, 0.1])  # 测试权重
w_test = w_test / np.sum(w_test)  # 归一化

print(f"初始权重: {w_test}")
print(f"初始方差: {portfolio_stats(mu_hat, Sigma_hat, w_test)['variance']:.6f}")

# 计算梯度
gradient = risk_gradient(Sigma_hat, w_test)
print(f"风险梯度: {gradient}")

# 沿负梯度方向更新权重
learning_rate = 0.01
w_new = w_test - learning_rate * gradient
w_new = w_new / np.sum(w_new)  # 重新归一化

print(f"更新后权重: {w_new}")
print(f"更新后方差: {portfolio_stats(mu_hat, Sigma_hat, w_new)['variance']:.6f}")

# 验证梯度下降是否降低了方差
old_var = portfolio_stats(mu_hat, Sigma_hat, w_test)['variance']
new_var = portfolio_stats(mu_hat, Sigma_hat, w_new)['variance']
print(f"方差变化: {old_var - new_var:.6f} ({'✅ 降低' if new_var < old_var else '❌ 增加'})")
````

````python
# 6.3 风险贡献可视化
print("\n📊 风险贡献可视化")
print("=" * 50)

# 计算不同投资组合的风险贡献
portfolios = {
    '等权重': w_equal,
    '全股票': w_stock,
    '测试组合': w_test
}

# 创建风险贡献对比图
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('[Day1] 不同投资组合的风险贡献分析', fontsize=16, fontweight='bold')

for i, (name, weights) in enumerate(portfolios.items()):
    stats = portfolio_stats(mu_hat, Sigma_hat, weights)
    
    # 绘制风险贡献条形图
    bars = axes[i].bar(ASSETS, stats['risk_contributions'], 
                      color=['red', 'orange', 'green', 'blue'], alpha=0.7,
                      edgecolor='black', linewidth=1)
    
    axes[i].set_title(f'{name}组合', fontweight='bold')
    axes[i].set_ylabel('风险贡献')
    axes[i].set_xlabel('资产类别')
    axes[i].grid(True, alpha=0.3)
    
    # 添加数值标签
    for bar, val in zip(bars, stats['risk_contributions']):
        height = bar.get_height()
        axes[i].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                    f'{val:.4f}', ha='center', va='bottom', fontweight='bold')
    
    # 添加组合统计信息
    axes[i].text(0.02, 0.98, f'期望收益: {stats["expected_return"]:.4f}\n标准差: {stats["std"]:.4f}\n夏普比率: {stats["sharpe_ratio"]:.4f}',
                transform=axes[i].transAxes, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.savefig(f'{OUTDIR}/risk_contributions.png', dpi=300, bbox_inches='tight')
plt.show()

# 分散化效果分析
print(f"📈 分散化效果分析:")
print(f"   等权重组合标准差: {stats_equal['std']:.6f}")
print(f"   全股票组合标准差: {stats_stock['std']:.6f}")
print(f"   分散化收益: {stats_stock['std'] - stats_equal['std']:.6f}")
print(f"   风险降低比例: {(stats_stock['std'] - stats_equal['std']) / stats_stock['std']:.1%}")

print(f"\n💡 关键洞察:")
print(f"   - 等权重组合通过分散化降低了风险")
print(f"   - 风险贡献显示各资产对组合风险的相对贡献")
print(f"   - 负相关资产(如Bond和Equity)有助于降低组合风险")
print(f"   - 梯度下降可以用于风险优化")
````

## Solutions（折叠区）

以下是所有练习的参考答案，点击展开查看详细解答。


````python
# 解决方案展示
from IPython.display import HTML, display

# Q1 解决方案
q1_solution = """
<details>
<summary><b>📖 Solution to Q1: 去均值函数实现</b></summary>

```python
def center(X: np.ndarray) -> np.ndarray:
    \"\"\"
    对矩阵X进行去均值处理
    
    Args:
        X: 输入矩阵 (T, n)
        
    Returns:
        去均值后的矩阵 (T, n)
    \"\"\"
    # 计算每列的均值
    mu = X.mean(axis=0)
    # 从每行减去对应列的均值
    return X - mu
```

**解释:**
- `X.mean(axis=0)` 计算每列的均值，得到形状为 (n,) 的向量
- `X - mu` 利用numpy的广播机制，从每行减去对应列的均值
- 这样处理后，每列的均值都变为0

</details>
"""

# Q2 解决方案
q2_solution = """
<details>
<summary><b>📖 Solution to Q2: 手动计算协方差矩阵</b></summary>

```python
def manual_covariance(X: np.ndarray) -> np.ndarray:
    \"\"\"
    手动计算协方差矩阵
    
    Args:
        X: 输入矩阵 (T, n)
        
    Returns:
        协方差矩阵 (n, n)
    \"\"\"
    # 1. 去均值
    X_centered = X - X.mean(axis=0)
    
    # 2. 计算 X^T * X
    XTX = X_centered.T @ X_centered
    
    # 3. 除以 (T-1)
    T = X.shape[0]
    return XTX / (T - 1)
```

**解释:**
- 协方差矩阵公式: Σ̂ = (1/(T-1)) * X^T * X，其中X是去均值矩阵
- 使用 `@` 运算符进行矩阵乘法
- 除以 (T-1) 而不是 T 是为了无偏估计

</details>
"""

# Q3 解决方案
q3_solution = """
<details>
<summary><b>📖 Solution to Q3: 协方差椭圆绘制</b></summary>

```python
def plot_covariance_ellipse_exercise(returns: pd.DataFrame, asset1: str, asset2: str) -> None:
    \"\"\"
    绘制协方差椭圆 (练习版本)
    \"\"\"
    # 1. 提取两资产数据
    data = returns[[asset1, asset2]].values
    mu = data.mean(axis=0)
    
    # 2. 计算协方差矩阵
    Sigma = np.cov(data.T)
    
    # 3. 特征分解
    eigenvals, eigenvecs = np.linalg.eigh(Sigma)
    eigenvals = np.maximum(eigenvals, 1e-10)  # 确保非负
    
    # 按特征值降序排列
    idx = np.argsort(eigenvals)[::-1]
    eigenvals = eigenvals[idx]
    eigenvecs = eigenvecs[:, idx]
    
    # 4. 绘制散点图
    plt.figure(figsize=(10, 8))
    plt.scatter(data[:, 0], data[:, 1], alpha=0.6, s=20, color='steelblue')
    plt.scatter(mu[0], mu[1], color='red', s=100, marker='x')
    
    # 5. 绘制椭圆
    angles = np.linspace(0, 2*np.pi, 100)
    for sigma in [1, 2]:
        a = sigma * np.sqrt(eigenvals[0])
        b = sigma * np.sqrt(eigenvals[1])
        
        ellipse_x = a * np.cos(angles)
        ellipse_y = b * np.sin(angles)
        
        # 旋转椭圆
        ellipse_rotated = eigenvecs @ np.array([ellipse_x, ellipse_y])
        ellipse_x_rot = ellipse_rotated[0] + mu[0]
        ellipse_y_rot = ellipse_rotated[1] + mu[1]
        
        plt.plot(ellipse_x_rot, ellipse_y_rot, '--', linewidth=2, 
                label=f'{σ}σ 椭圆')
    
    plt.xlabel(asset1)
    plt.ylabel(asset2)
    plt.title(f'{asset1} vs {asset2} 协方差椭圆')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.axis('equal')
    plt.show()
```

**解释:**
- 椭圆的长轴和短轴长度与特征值的平方根成比例
- 椭圆的旋转角度由特征向量决定
- 1σ椭圆应包含约68%的数据点，2σ椭圆应包含约95%的数据点

</details>
"""

# Q4 解决方案
q4_solution = """
<details>
<summary><b>📖 Solution to Q4: 组合方差与梯度验证</b></summary>

```python
def gradient_descent_step(w: np.ndarray, Sigma: np.ndarray, learning_rate: float = 0.01) -> np.ndarray:
    \"\"\"
    执行一步梯度下降
    \"\"\"
    # 1. 计算梯度
    grad = risk_gradient(Sigma, w)
    
    # 2. 更新权重
    w_new = w - learning_rate * grad
    
    # 3. 归一化
    w_new = w_new / np.sum(w_new)
    
    return w_new
```

**解释:**
- 梯度指向风险增加最快的方向
- 负梯度方向是风险降低最快的方向
- 学习率控制步长，太小收敛慢，太大可能发散
- 归一化确保权重和为1

</details>
"""

# Q5 解决方案
q5_solution = """
<details>
<summary><b>📖 Solution to Q5: 主成分重构协方差矩阵</b></summary>

```python
def reconstruct_covariance(Sigma: np.ndarray, k: int) -> np.ndarray:
    \"\"\"
    用前k个主成分重构协方差矩阵
    \"\"\"
    # 1. 特征分解
    eigenvals, eigenvecs = np.linalg.eigh(Sigma)
    
    # 2. 按特征值降序排列
    idx = np.argsort(eigenvals)[::-1]
    eigenvals = eigenvals[idx]
    eigenvecs = eigenvecs[:, idx]
    
    # 3. 取前k个
    eigenvals_k = eigenvals[:k]
    eigenvecs_k = eigenvecs[:, :k]
    
    # 4. 重构: Σ_k = V_k @ diag(λ_k) @ V_k^T
    return eigenvecs_k @ np.diag(eigenvals_k) @ eigenvecs_k.T

def frobenius_norm_error(Sigma_original: np.ndarray, Sigma_reconstructed: np.ndarray) -> float:
    \"\"\"
    计算Frobenius范数误差
    \"\"\"
    diff = Sigma_original - Sigma_reconstructed
    return np.sqrt(np.sum(diff**2))
```

**解释:**
- 主成分重构是降维技术，保留最重要的信息
- Frobenius范数衡量矩阵间的"距离"
- 重构误差随k增加而减小，但k=n时误差为0

</details>
"""

# 显示所有解决方案
display(HTML(q1_solution))
display(HTML(q2_solution))
display(HTML(q3_solution))
display(HTML(q4_solution))
display(HTML(q5_solution))
````
